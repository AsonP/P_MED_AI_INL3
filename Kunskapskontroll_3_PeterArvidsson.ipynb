<strong>Genomsnittligt husvärde och medianvärde baserat på närhet till hav</strong>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Läs in data från en CSV-fil 
#df = pd.read_csv("housing.csv", sep=';')
#df.groupby('ocean_proximity')['median_house_value'].mean()

# Läs in data från CSV
df = pd.read_csv("housing.csv", sep=';')

# Beräkna medelvärden och sortera
summary = df.groupby('ocean_proximity')['median_house_value'].mean().sort_values(ascending=False)

# Formatera tabellen och skriv ut
formatted_summary = summary.round(2)
print("Genomsnittligt husvärde baserat på närhet till hav:")
print(formatted_summary)
print()

# Visualisera med en stapelgraf
plt.figure(figsize=(8, 5))
formatted_summary.plot(kind='bar', color='skyblue', alpha=0.8, edgecolor='black')
plt.title('Genomsnittligt Medianvärde för hus med närhet till havet', fontsize=14)
plt.ylabel('Genomsnittligt Husvärde', fontsize=12)
plt.xlabel('Närhet till Hav (Ocean Proximity)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

<strong>Relation mellan Median Inkomst och Median Husvärde</strong>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Läs in data från CSV
df = pd.read_csv("housing.csv", sep=';')

x = df['median_income']
y = df['median_house_value']

# Beräkna linjär regression
m, b = np.polyfit(x, y, 1)  # Lutning (m) och intercept (b)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.5, color='blue', label='Data')
plt.plot(x, m * x + b, color='red', label=f'Trendlinje: y = {m:.2f}x + {b:.2f}')
plt.xlabel('Median Income', fontsize=12)
plt.ylabel('Median House Value', fontsize=12)
plt.title('Relation mellan Median Income och Median House Value', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

<strong> Husvärden: Fördelning, Medel och Median</strong>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Ladda datasetet
housing_data = pd.read_csv("housing.csv", sep=';')

# Data: Median Husvärde
house_values = housing_data['median_house_value'].values

# Beräkningar med NumPy
mean_value = np.mean(house_values)
median_value = np.median(house_values)
std_deviation = np.std(house_values)

# Visualisering
plt.figure(figsize=(10, 6))
plt.hist(house_values, bins=50, color='skyblue', alpha=0.7, label='Husvärden')
plt.axvline(mean_value, color='red', linestyle='--', label=f'Medelvärde: ${mean_value:,.0f}')
plt.axvline(median_value, color='green', linestyle='-', label=f'Median: ${median_value:,.0f}')
plt.title('Husvärden: Fördelning, Medel och Median', fontsize=16)
plt.xlabel('Husvärde ($)', fontsize=12)
plt.ylabel('Antal', fontsize=12)
plt.legend()
plt.show()
```

<strong>Identifiera de dyraste husen (Räkna hur många hus det finns i varje kategori av 'ocean_proximity')</strong>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Ladda datasetet
housing_data = pd.read_csv("housing.csv", sep=';')

# Filtrera dyra hus
expensive_houses = housing_data[housing_data['median_house_value'] > 400000]

# Räkna hur många hus det finns i varje kategori av 'ocean_proximity'
categories = expensive_houses['ocean_proximity'].value_counts()

# Visualisering
plt.figure(figsize=(8, 5))
plt.bar(categories.index, categories.values, color='orange', alpha=0.7)
plt.title('Antal dyra hus per närhet till havet', fontsize=16)
plt.xlabel('Närhet till havet', fontsize=12)
plt.ylabel('Antal hus', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

<strong>Korrelation mellan medianinkomst och husvärde</strong>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Läs in data
df = pd.read_csv("housing.csv", sep=';')

# Korrelation mellan medianinkomst och husvärde
correlation_income_value = df[['median_income', 'median_house_value']].corr().iloc[0, 1]
print(f"Korrelation mellan medianinkomst och husvärde: {correlation_income_value:.2f}")

# Genomsnittligt husvärde per kategori för närhet till hav
mean_values_ocean = df.groupby('ocean_proximity')['median_house_value'].mean().sort_values()

# Scatterplot mellan medianinkomst och husvärde
plt.figure(figsize=(10, 6))
plt.scatter(df['median_income'], df['median_house_value'], alpha=0.5, c=df['median_income'], cmap='viridis', edgecolor='k')
plt.colorbar(label='Median Income')
plt.title('Relation mellan Median Income och Median House Value', fontsize=14)
plt.xlabel('Median Income ($)', fontsize=12)
plt.ylabel('Median House Value ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Dyra hus: Hus nära havet (t.ex. <1H OCEAN, NEAR BAY) och områden med hög medianinkomst är generellt dyrare. En hög korrelation mellan medianinkomst och median_house_value tyder på att inkomstnivån är en stark drivkraft.

Billiga hus: Hus längre från havet (t.ex. INLAND) och områden med låg medianinkomst är billigare. Stora befolkningar per område kan också påverka värdet negativt (trängsel och lägre efterfrågan).
